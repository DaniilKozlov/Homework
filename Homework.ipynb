```python
df=pd.read_excel('data/bank_account_data.xlsx')
display(df) # проверяем вывод данных в таблице
df.describe() # проверяем вывод данных о таблице
```

```python
#Задание 1 - посчитать показатели describe
descr=ps.sqldf('''
SELECT 
    COUNT(cust_id) cust_id, 
    COUNT(purchase) purchase, 
    COUNT(Cheq_Bal) cheq_bal, 
    COUNT(sav_bal) sav_bal,  
    COUNT(debt_bal) debt_bal,  
    COUNT(joint_acct) joint_acct,  
    COUNT(age) age 
FROM df
UNION ALL
SELECT 
    AVG(cust_id), 
    AVG(purchase), 
    AVG(Cheq_Bal), 
    AVG(sav_bal),  
    AVG(debt_bal),  
    AVG(joint_acct),  
    AVG(age) 
FROM df
UNION ALL
SELECT 
    MAX(cust_id), 
    MAX(purchase), 
    MAX(Cheq_Bal), 
    MAX(sav_bal),  
    MAX(debt_bal),  
    MAX(joint_acct),  
    MAX(age) 
FROM df
UNION ALL
SELECT 
    MIN(cust_id), 
    MIN(purchase), 
    MIN(Cheq_Bal), 
    MIN(sav_bal),  
    MIN(debt_bal),  
    MIN(joint_acct),  
    MIN(age) 
FROM df;
''')

display(descr)
print('0 - Count, 1 - mean, 2 - Max, 3 - Min')
print('Функции рассчета std, квантилей в pandasql отсутствуют')
```

```python
#Задание 2 - сумма по всем балансам
sum_bal=ps.sqldf('''
SELECT 
    SUM(cheq_bal) Cheq_Bal_sum, 
    SUM(Sav_Bal) Sav_Bal_sum, 
    SUM(Debt_Bal) Debt_Bal_sum
FROM df;
''')

display(sum_bal)
```

```python
#Задание 3 - только клиенты с ипотекой
clients_with_I=ps.sqldf('''
SELECT * 
FROM df
    WHERE purchase >0;
''')

display(clients_with_I)
```

```python
#Задание 4 - совокупный объем средств с группировкой по критерию наличия ипотеки
sum_bal_with_group=ps.sqldf('''
SELECT 
    purchase, 
    SUM(Cheq_Bal) Cheq_bal_sum_gr, 
    SUM(Sav_Bal) Sav_bal_sum_gr, 
    SUM(Debt_Bal) Debt_bal_sum_gr
FROM df
GROUP BY purchase;
''')

display(sum_bal_with_group)
```

```python
#Задание 5 - найти разницу между средними показателями всех счетов у клиентов с ипотекой и без ипотеки
dif=ps.sqldf('''
SELECT DISTINCT 
    (SELECT avg(cheq_bal) FROM df WHERE purchase=1 GROUP BY purchase) - (SELECT avg(cheq_bal) FROM df WHERE purchase=0 GROUP BY purchase) AS Cheq_dif,
    (SELECT avg(sav_bal) FROM df WHERE purchase=1 GROUP BY purchase) - (SELECT avg(sav_bal) FROM df WHERE purchase=0 GROUP BY purchase) AS Sav_dif,
    (SELECT avg(debt_bal) FROM df WHERE purchase=1 GROUP BY purchase) - (SELECT avg(debt_bal) FROM df WHERE purchase=0 GROUP BY purchase) AS Debt_dif
FROM df
    GROUP BY purchase;
''')

display(dif)
print('Разница между средними величинами по счетам клиентов с ипотекой и без ипотеки: есть')
```

```python
#Задание 6 - самый возрастной получатель ипотеки
max_age=ps.sqldf('''
SELECT *
FROM df
    WHERE age=(SELECT max(age)from df);
''')

display(max_age)
print('Таких оказалось несколько')
```

```python
#Задание 7 - самый первый получатель ипотеки 
first_client=ps.sqldf('''
SELECT *
FROM df
    WHERE date=(SELECT min(date) FROM df);
''')

display(first_client)
```
